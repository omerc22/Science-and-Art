```python
import torch
from diffusers import StableDiffusionPipeline
import os

# Create the outputs directory if it doesn't exist
output_dir = "/outputs"
os.makedirs(output_dir, exist_ok=True)

# Load the pre-trained Stable Diffusion model
# You might need to accept the user license agreement for this model on Hugging Face.
# If you encounter issues, visit the model page (e.g., runwayml/stable-diffusion-v1-5) and accept the terms.
model_id = "runwayml/stable-diffusion-v1-5"
pipeline = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipeline.to("cuda")
```

# **1**

```python
# Define your text prompt
prompt = "Neon-glowing tramway, rain-soaked Taksim Square, cyberpunk Istanbul, art-deco architecture, dim lights, 8k, photorealistic"

# Generate the image
image = pipeline(prompt).images[0]

# Define the filename and save the image
image_filename = os.path.join(output_dir, "1.png")
image.save(image_filename)

print(f"Image saved to {image_filename}")

# Display the image (optional)
from IPython.display import display
display(image)
```
